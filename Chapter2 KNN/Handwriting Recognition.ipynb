为了使用前面的分类器，我们把$ 32 \times 32 $的二进制图像矩阵转化为$ 1\times1024 $的向量

```
def img2vector(filename):
    returnVect = zeros((1,1024))
    fr = open(filename)
    for i in range(32):
        lineStr = fr.readline() 每次读取一行
        for j in range(32):
            returnVect[0,32*i+j] = int(lineStr[j]) 添加到向量中
    return returnVect
```
作用：将图像转化为向量

方法：该函数创建$ 1\times1024 $的numpy数组，然后打开给定的文件，循环读出文件前32行，并将每行的头32个字符值储存在Numpy数组中，最后返回数组。

这种全部式的添加保证了，数据的完整性。对于结果的准确性至关重要。
